```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
# get dataset
house_data = sklearn.datasets.fetch_california_housing()
type(house_data)
```

```python
print(house_data)
```

```python
# convert Bunch to DataFrame -> 1st step: move 'data' to DataFrame (as X) -> 2nd step: add 'target' to the same DataFrame (as Y)
house_price_data = pd.DataFrame(house_data.data, columns=house_data.feature_names)
house_price_data.head()
```

```python
house_price_data['price']=house_data.target
house_price_data.head()
```

```python
house_price_data.isna().sum()
```

```python
house_price_data.describe()
```

```python
corelation = house_price_data.corr()
```

```python
# construct a heatmap to show correlation
sns.heatmap(corelation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':6}, cmap='Oranges')
```

```python
# Split DataFrame into X and Y
X = house_price_data.drop(["price"], axis=1)
Y = house_price_data['price']
```

```python
X.head()
```

```python
Y.head()
```

```python
# split data into train / test data
trainX, testX, trainY, testY = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, trainX.shape, testX.shape)
```

```python
# train model
model = XGBRegressor()
model.fit(trainX, trainY)
```

```python
predict_y = model.predict(trainX)
```

```python
predict_y
```

```python
# for regression, use r2-score and mean_absolute_error to evaluate the model
r2score = metrics.r2_score(trainY, predict_y)
r2score
```

```python
maerror= metrics.mean_absolute_error(trainY, predict_y)
maerror
```

```python
# also check both for test data
predict_y_test = model.predict(testX)
```

```python
r2score_test = metrics.r2_score(testY, predict_y_test)
r2score_test
```

```python
maerror_test = metrics.mean_absolute_error(testY, predict_y_test)
maerror_test
```

Both test and train dataset has pretty low error evaluation, which is good for the model

**Step 1: Handle Missing Values**
1. imputation ( central tendencies : mean, median or mode )
2. drop rows : dataset.dropna(how='any')

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_set = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Placement_Dataset.csv')
```

```python
data_set.head()
```

```python
data_set.shape
```

```python
data_set.isna().sum()  # salary has 67 null values
```

```python
# imputation - use statistical values(mean, median or mode => central tendencies ) to replace those missing values
# analysis 'salary' distribution:
# if normal distribution, use mean, else with skewed distribution( outliers ) use median or mode for replacement

fig, ax = plt.subplots()
ax = sns.histplot(data_set.salary, bins=40) # skewed with mode reside in
# add annotations
for p in ax.patches:
    color = p.get_facecolor()
    ax.annotate(f"{p.get_height():.0f}\n", (p.get_x() + p.get_width() / 2, p.get_height()),
                ha="center", va="center", color=color, fontweight="bold")
```

```python
# replace all missing salaries with medians => inplace=True to make changes on the original dataset
data_set['salary'].fillna(data_set.salary.median(), inplace=True)
```

**Step 4: Handle imbalanced Dataset**
such as 1000 M and only 100 B inside a dataset col.The imblanace has to be removed before feed it into a model for training
. Using 'Under Sampling' technique: take similar amount from big group (sample()), concat() it with the smaller group to form randomly selected new dataset.

```python
import pandas as pd
credit = pd.read_csv("/content/drive/MyDrive/credit_data.csv")
credit.tail()
```

```python
# check if any missing values
credit.isna().sum()  # no missing values
```

```python
# check if need to standardization
from sklearn.preprocessing import StandardScaler
credit.std() #
```

```python
credit.Class.value_counts() # 0    284315   vs. 1       492  => imbalanced
```

```python
# seperate legit(0) and fraudulent(1) transactions into 'legit' and 'fraudulent' variables
# conditionally select data -> or alternatively use isin()
legit = credit[credit.Class== 0]
legit.shape
```

```python
fraudulent = credit[credit.Class==1]
fraudulent.shape
```

```python
# practice with conditionally select data
times = credit[credit['Time'] > 100000]
times.shape
```

```python
(credit['Amount']).max()
```

```python
# under-sampling -> sample the same (similar) amount of data
legit_sample = legit.sample(n= 492) # 492 randomly chosen rows with legit class of 0
legit_sample.shape
```

```python
# add the legit_sample to fraudulent data
balanced_credit = pd.concat((legit_sample, fraudulent), axis=0) # add as rows
balanced_credit.shape
```

```python
balanced_credit.head()  # the first col is randomly selected index #
```

```python
balanced_credit.tail()
```

```python
# check new data balance
balanced_credit.Class.value_counts()
```
